```python
import pandas as pd
```

```python
df = pd.read_csv("./metro.csv", delimiter=';')
```

```python
df.columns
```

```python
df = df[['Réseau', 'Station', 'Trafic', 'Correspondance_1',
       'Correspondance_2', 'Correspondance_3', 'Correspondance_4',
       'Correspondance_5', 'Ville', 'Arrondissement pour Paris']]
```

```python
df = df.rename({"Arrondissement pour Paris": "arrondissement"}, axis=1)
df = df.rename(lambda c: c.lower().replace(" ", "_"), axis=1)
df = df.apply(lambda x: x.astype(str).str.title())
df
```

```python
df = df.astype({'arrondissement': float, "correspondance_4": float, "correspondance_5": float})
```

```python
from sqlalchemy import create_engine    

engine = create_engine('sqlite://', echo=False)
df.to_sql(name="metro", con=engine, index=False)

with engine.connect() as conn:
    with open("metro.sql", "w") as f:
        for line in conn.connection.iterdump():
            f.write(line)
            f.write('\n')
```
