## Exercice 7

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2*np.sin(1/x)
def fp(x):
    return 2*x*np.sin(1/x) - np.cos(1/x)

def g(x):
    return f(x) + 2*x
def gp(x):
    return fp(x) + 2

def h(x):
    return f(x) + x
def hp(x):
    return fp(x) + 1
```

```python
X = np.arange(-10, 10, 0.01)
Y = [fp(x) for x in X]
plt.plot(X, Y)
plt.show()
```

```python
X = np.arange(-1, 1, 0.01)
Y_G = [g(x) for x in X]
Y_GP = [gp(x) for x in X]
plt.plot(X, Y_G)
plt.plot(X, Y_GP)
plt.show()
```

```python
Y_H = [h(x) for x in X]
Y_HP = [hp(x) for x in X]
plt.plot(X, Y_H)
plt.plot(X, Y_HP)
plt.legend(['h', 'h\''])
plt.show()
```

```python
N = range(18)
Y_Hp = []
for n in N:
    n = 30 + 10*n
    Un = 2*n*np.pi - 0.5/n
    Y_Hp.append(hp(1/Un))
plt.plot(N, Y_Hp)
plt.show()
```

## Exercice 8

```python
def D1(n, a, b):
    if n == 2:
        return a**2 -2*b
    if n == 3:
        return a**3 - 3*a*b
    return a*D1(n-1, a, b) - b*D1(n-2, a, b)
```

```python
b = 3
for n in range(3, 6):
    X = np.arange(-2*b**.5, 2*b**.5, 0.01)
    Y = [D1(n, x, b) for x in X]
    plt.plot(X, Y)
    plt.show()
# a -> Dn(a, b) semble avoir n racines simples
```
