## Exercice 2

```python
import numpy as np
import numpy.random as rd
import matplotlib.pyplot as plt
```

```python
def X(k, n):
    return [rd.randint(1, n + 1) for _ in range(k)]
```

```python
# 2)
def S(k):
    Xk = X(k, 10)
    L = [0]*10 # L[i] = 1 si i+1 est dans Xk
    for x in Xk:
        L[x - 1] = 1
    return sum(L)

def moyenne(k):
    s = 0
    for _ in range(1000):
        s += S(k)
    return s/1000

print(moyenne(10))
```

```python
# 4)
x = range(1, 21)
plt.plot(x, [np.log(1 - moyenne(k)/10) for k in x])
plt.grid()
plt.show() 
print(np.log(1 - moyenne(1)/10))
print(np.log(1 - moyenne(20)/10))
# conjecture : np.log(1 - moyenne(k)/10) = ak + b soit moyenne(k) = 10(1 - exp(ak + b)) avec a < 0
```

## Exercice 3

```python
import numpy as np
import numpy.random as rd

# 3)a)
def diag(n):
    d = np.zeros((n, n))
    for i in range(n):
        d[i][i] = rd.random()
    return d

diag(5)
```

```python
# 3)b)
def Matrice(A, B):
    return A.dot(B) - B.dot(A)

Matrice(diag(5), rd.rand(5, 5)) # les coefficients diagonaux sont nuls
```
