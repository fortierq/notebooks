# Exercice 4

```python
import matplotlib.pyplot as plt
import scipy.optimize as resol

def x(n):
    def f(x):
        s = 0
        for k in range(1, n + 1):
            s += x**k/k
        return s - 1
    return resol.fsolve(f, 0.5)[0]

# 2)
N = range(1, 51)
plt.plot(N, [x(n) for n in N])
plt.show() # conjecture : xn -> ~0.63 
```

## Exercice 5

```python
import numpy as np

def U(n):
    m  = 0
    for i in range(1000):
        M = np.random.randint(0, 2, (n, n))
        m = max(m, np.linalg.det(M))
    return m

def V(n):
    m  = 0
    for i in range(1000):
        M = np.random.random((n, n))
        m = max(m, np.linalg.det(M))
    return m

[U(n) for n in range(11)], [V(n) for n in range(11)]
# les valeurs de U(n) et V(n) semblent différentes...
```

```python
def M(n, p):
    X = []
    for i in range(n):
        if rd.random() < p:
            X.append(1)
        else:
            X.append(0)
    m = np.zeros((n, n))
    for i in range(n):
        for j in range(n):
            m[i][j] = X[i]*X[j]
    return m

k = 0
for i in range(1000):
    if np.trace(M(5, 0.5)) <= 1:
        k += 1
k/1000
```

## Exercice 6

```python
# Pour que a/n <= 10^-3, il faut que n >= a * 10^3

def S(a):
    s = 0
    n = 1000*a
    for k in range(1, n + 1):
        s += a/(k**2 + a**2)
    return s

S(1), S(5)
```

```python
X = range(1, 51)
Y = [S(a) for a in X]
plt.plot(X, Y)
plt.show() # conjecture : S(a) -> pi/2
```
